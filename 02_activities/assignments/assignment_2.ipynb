# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Imports (put at top of notebook)
%load_ext dotenv
%dotenv

import os, sys
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

import pickle
import shap   # pip install shap if missing
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
# header=0 → tells pandas that row 0 (the first line of the CSV) contains column names.
# names=columns → forces pandas to ignore those names and instead use the ones I supplied in columns.
fires_dt.info()
```

# Get X and Y

Create the features data frame and target data.

```python
X = fires_dt[['coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']]
# x = features

Y = fires_dt['area']
# y = target


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
num_features = ['coord_x', 'coord_y', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
cat_features = ['month', 'day']
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
preproc1 = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ],
    remainder='drop'
)
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
from sklearn.pipeline import Pipeline as SKPipeline
numeric_pipeline = SKPipeline([
    ('scaler', StandardScaler()),
    ('power', PowerTransformer())   # the non-linear transform
])

preproc2 = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, num_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
    ],
    remainder='drop'
)
```

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
# Pipeline A = preproc1 + baseline
from sklearn.linear_model import Lasso

pipeline_A = Pipeline(
    [
        ('preproc', preproc1), 
        ('regressor', Lasso())
    ]
)

pipeline_A

# lasso regression
# https://uoft-dsi-certificates.slack.com/archives/C095R8YVACR/p1759415861640369?thread_ts=1759413616.796039&cid=C095R8YVACR
```

```python
# Pipeline B = preproc2 + baseline

from sklearn.neighbors import KNeighborsRegressor

pipeline_B = Pipeline([
    ('preproc', preproc2),
    ('regressor', KNeighborsRegressor())
])

pipeline_B
```

```python
# Pipeline C = preproc1 + advanced model
from sklearn.ensemble import GradientBoostingRegressor

pipeline_C = Pipeline([
    ('preproc', preproc1),
    ('regressor', GradientBoostingRegressor(random_state=42))
])

pipeline_C
```

```python
# Pipeline D = preproc2 + advanced model

from sklearn.ensemble import GradientBoostingRegressor

pipeline_D = Pipeline([
    ('preproc', preproc2),
    ('regressor', GradientBoostingRegressor(random_state=42))
])

pipeline_D
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
pipeline_A.get_params()
```

```python
pipeline_C.get_params()
```

```python
# in labs: 05_hyperparams
param_grid_A = {
    'regressor__alpha': [0.001, 0.01, 0.1, 1.0],
    'regressor__max_iter': [1000, 5000, 10000]
}

scoring = ['r2', 'neg_mean_squared_error', 'neg_mean_absolute_error']

grid_A = GridSearchCV(
    estimator=pipeline_A,
    param_grid=param_grid_A,
    scoring=scoring,
    cv=5,
    refit='r2',   # refit best model using R^2
    n_jobs=-1,
    verbose=1
)

grid_A.fit(X_train, Y_train)
print("Best R2:", grid_A.best_score_)
print("Best params:", grid_A.best_params_)
```

```python
param_grid_B = {
    'regressor__n_neighbors': [3, 5, 7, 9],
    'regressor__weights': ['uniform', 'distance'],
    'regressor__p': [1, 2]  # Manhattan or Euclidean
}

grid_B = GridSearchCV(
    estimator=pipeline_B,
    param_grid=param_grid_B,
    scoring=scoring,
    cv=5,
    refit='r2',
    n_jobs=-1,
    verbose=1
)

grid_B.fit(X_train, Y_train)
```

```python
param_grid_C = {
    'regressor__n_estimators': [100, 200],
    'regressor__learning_rate': [0.05, 0.1],
    'regressor__max_depth': [3, 5],
    'regressor__min_samples_split': [2, 5],
    'regressor__min_samples_leaf': [1, 2],
    'regressor__subsample': [0.8, 1.0]
}

grid_C = GridSearchCV(
    estimator=pipeline_C,
    param_grid=param_grid_C,
    scoring=scoring,
    cv=5,
    refit='r2',
    n_jobs=-1,
    verbose=1
)

grid_C.fit(X_train, Y_train)
```

```python
param_grid_D = param_grid_C  # same grid as pipeline C

grid_D = GridSearchCV(
    estimator=pipeline_D,
    param_grid=param_grid_D,
    scoring=scoring,
    cv=5,
    refit='r2',
    n_jobs=-1,
    verbose=1
)

grid_D.fit(X_train, Y_train)
```
